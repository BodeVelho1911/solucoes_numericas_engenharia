```python
import numpy as np

# Exemplo de implementação do método de Gauss-Jacobi em Python para determinar coeficientes de sistemas lineares
# Inputs
A = np.array( [ [3,2],[1,-2] ] )
b = np.array( [ [5,-1] ] ).T
x0 = np.array([ [0,0] ]).T
tol = 1e-7

# Cálculo da matriz [C] e vetor [d]
diagonal = np.diag(A).reshape(2, 1)
d = b / diagonal
C = -A / diagonal + np.eye(2)

# Método de Gauss-Jacobi
x1 = np.dot(C, x0) + d
iter_num = 0
while np.max(abs(x1 - x0)) > tol:
  x0 = x1
  x1 = np.dot(C, x0) + d
  iter_num += 1

# Output
print(f'O solução aproximada com precisão de {tol} é:\n {x1}')
print(f'Número de iterações: {iter_num}')
```
