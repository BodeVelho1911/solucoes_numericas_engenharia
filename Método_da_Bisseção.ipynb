# Bisection method flowchart

```python
import numpy as np

# Função para calcular f(x) = x * e^x - t, onde x e t são variáveis
# Substitua 'x * np.exp(x) - t' pela função que deseja achar raízes
def f(x, t):
    y = x * np.exp(x) - t
    return y

# Função para método da bisseção
def bisection_method(t, a, b, eps):
    # Verifica se há uma raiz no intervalo [a, b]
    if f(a, t) * f(b, t) >= 0:
        print("O intervalo [a, b] não contém uma raiz.")
        return None, None

    cont_it = 0
    x = (a + b) / 2
    # Executa o método até atingir a precisão desejada
    while abs(f(x, t)) > eps:
        if f(a, t) * f(x, t) < 0:
            b = x
        else:
            a = x
        x = (a + b) / 2
        cont_it += 1

    return x, cont_it

# Valores de t a serem testados
t_values = [1, 10, 100]
a = 0  # Limite inferior do intervalo
eps = 1e-6  # Precisão desejada

for t in t_values:
    b = t  # Limite superior do intervalo
    raiz, iteracoes = bisection_method(t, a, b, eps)
    if raiz is not None:
        print(f"Para t = {t}:")
        print(f"Raiz aproximada: {raiz}")
        print(f"Número de iterações: {iteracoes}")
    print()
```

```python
#Inputs do código
a = float(input('Entre com o valor do limite inferior do intervalo de procura da raiz: '))
b = float(input('Entre com o valor do limite superior do intervalo de procura da raiz: '))
eps = float(input('Entre com a precisão desejada: '))

# Função para calcular f(x) = e ** x - sen(x) ==>
# Caso deseje, substitua a função y = np.log(x) + 2 * x pela qual deseja calcular as raízes
def f(x):
  y = np.log(x) + 2 * x
  return y

# Utilizando o método da bisseção
cont_it = 0
x = (a + b) / 2
while abs(f(x)) > eps:
  if f(a) * f(x) < 0:
    b = x
  else:
    a = x
  x = (a + b) / 2
  cont_it += 1

#Output do código---------------------------------------------------------------
print('A raiz aproximada é: {}'.format(x))
print('O número de iterações realizadas foram: {}'.format(cont_it))
```
