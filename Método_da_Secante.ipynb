```python
#Importação da biblioteca numpy
import numpy as np

#Inputs do código (entrada pelo teclado)----------------------------------------
x0 = float(input('Entre com um valor para x0: '))
x1 = float(input('Entre com um valor para x1: '))
eps = float(input('Entre com a precisão desejada: '))
def f(x):
  y = x ** 2 - 3
  return y

#Método da Secante-------------------------------------------------------------

if f(x1) - f(x0) != 0:
  n = 0
  x2 = x1 - ((f(x1) * (x1 - x0)) / (f(x1) - f(x0)))
  while abs(f(x2)) > eps:
    x0 = x1
    x1 = x2
    x2 = x1 - ((f(x1) * (x1 - x0)) / (f(x1) - f(x0)))
    print(x2)
    n = n + 1

  #Output do código---------------------------------------------------------------
  print(f'A raiz aproximada é: {x2:.5f}')
  print(f'O número de iterações realizadas foram: {n}')
else:
  print('Divisão por zero!! Digite novos valores')

```

```python
# Exercício de CVT
#Importação da biblioteca numpy
import numpy as np

#Inputs do código (entrada pelo teclado)----------------------------------------
x0 = float(input('Entre com um valor para x0: '))
x1 = float(input('Entre com um valor para x1: '))
eps = float(input('Entre com a precisão desejada: '))
def f(r2):
  alfa = np.arcsin((r2 - 0.1) / 0.5)
  beta1 = np.pi - 2 * alfa
  beta2 = np.pi + 2 * alfa
  y = 0.1 * beta1 + r2 * beta2 + 2 * np.sqrt(0.1 ** 2 - (r2 - 0.1) ** 2) - 2
  return y

#Método da Secante-------------------------------------------------------------

if f(x1) - f(x0) != 0:
  n = 0
  x2 = x1 - ((f(x1) * (x1 - x0)) / (f(x1) - f(x0)))
  while abs(f(x2)) > eps:
    x0 = x1
    x1 = x2
    x2 = x1 - ((f(x1) * (x1 - x0)) / (f(x1) - f(x0)))
    print(x2)
    n = n + 1

  #Output do código---------------------------------------------------------------
  print(f'A raiz aproximada é: {x2:.5f}')
  print(f'O número de iterações realizadas foram: {n}')
else:
  print('Divisão por zero!! Digite novos valores')

```
